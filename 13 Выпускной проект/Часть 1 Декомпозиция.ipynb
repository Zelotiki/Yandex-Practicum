# Сегментация пользователей по потреблению 

## Описание проекта 

Проанализировать клиентов регионального банка и сегментировать пользователей по количеству потребляемых продуктов.
    
В нашем распоряжении датасет bank_scrooge.csv, который содержит информацию о клиентах банка «Метанпром». Банк расположен в Ярославле и городах области: Ростов Великий и Рыбинск.

## Цели исследования:

##### Уменьшение оттока клиентов банка благодаря анализу поведения клиентов, который включает в себя:

- разделение пользователей на основе данных о количестве потребляемых продуктов
- исследование отличительных черт получившихся групп и сформировать более компактные сегменты, в которых эти черты будут более выражены
- исследование факторов, усиливающих отток клиентов и удерживающих клиентов
- проверка гипотезы, о различии возраста между клиентами, кто пользуются двумя продуктами, и теми, кто пользуются одним

## План проработки проекта: 

### 1. Сбор данных:

- Настройка сбора данных
- Анализ датасета
- Получение информации по типам, полноте и пропускам.

### 2. Предобработка данных:

- Проработка столбцов по единому регистру 
- Поиск и работа с дубликатами
- Выявление столбцов с пропусками и решения их заполнения/удаления
- Подготовка данных (переводим категориальные переменные в числовые), стандартизация данных

### 3. Исследовательский анализ данных:

- Наблюдение распределения признаков по оттока клиентов, выявление тенденций для удержания
- Исследование коррелирующих признаков с помощью Heatmap
- Создание визуализаций зависимости между разными признаками 
- Распределение признаков в разрезе филиалов банка
- Распределение признаков в разрезе потребления

### 4. Проверка гипотез

- Проверка гипотезы о различии возраста между клиентами, кто пользуются двумя продуктами банка, и теми, кто пользуются одним

#### Этапы проведения тестирования:
1. Группировка данных по выдвинутой гипотезе
2. Строим точечный график возраста/баланса клиента в группах, наблюдение за выбросами
3. Подготовка данных к тестированию
4. Выбор метода для проведения теста и тестирование
5. Выводы полученных результатов

### 5. Сегментация групп

- Сегментация клиентов с целью определения факторов оттока
- Описание выявленых сегментов 
- Составяление портретов и формирование рекомендации
